# Draw a visualization of the cities

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
city_pos = np.genfromtxt("att48.tsp", dtype=int, delimiter=' ', skip_header=6, skip_footer=1, usecols=(1, 2))

fig, ax = plt.subplots(figsize=(5, 5))
ax.scatter(city_pos[:,0], city_pos[:,1])
print(city_pos)
print(len(city_pos))
```

```python
circuit = list(range(len(city_pos))) + [1]

def drawLine(c):
    for i in range(len(c)-1):
        ax = plt.plot([city_pos[c[i]-1,0], city_pos[c[i+1]-1,0]], [city_pos[c[i]-1,1], city_pos[c[i+1]-1,1]], 'bo-')
    for i, val in enumerate(c):
        plt.annotate(val, xy=(city_pos[c[i]-1,0],city_pos[c[i]-1,1]), xytext=(5,0), textcoords='offset points')


drawLine(circuit)
```

```python
circuit2 = list(range(len(city_pos))) + [1]

def dist(c):
    res = 0
    def d(x1, x2, y1, y2):
        return np.sqrt((x1-x2)**2 + (y1-y2)**2)
    for i in range(len(c)-1):
        res += d(city_pos[c[i]-1,0], city_pos[c[i+1]-1,0], city_pos[c[i]-1,1], city_pos[c[i+1]-1,1])
    return res

dist(circuit2)
```
